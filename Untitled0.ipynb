```python
# import rasterio
# from rasterio.windows import Window
# import numpy as np
# import glob
# import os
# import random
# from tqdm import tqdm
# import h5py
# import shutil
# import gc

# CHIP_SIZE = 256
# STRIDE = 128
# DISTRICT_TARGET = "Ahmadabad"
# PROCESS_CHUNK_ROW_SIZE = 1024
# INPUT_FOLDER = "/content/drive/MyDrive/Gujarat_Satellite_Data"
# TEMP_OUTPUT_DIR = "/content/temp_shards"
# DRIVE_BACKUP_DIR = "/content/drive/MyDrive/ahmedabad_hdf5_shards"

# os.makedirs(TEMP_OUTPUT_DIR, exist_ok=True)
# os.makedirs(DRIVE_BACKUP_DIR, exist_ok=True)

# def calculate_index(b1, b2):
#     with np.errstate(divide='ignore', invalid='ignore'):
#         idx = (b1 - b2) / (b1 + b2 + 1e-8)
#     return np.nan_to_num(idx, nan=-1.0)

# class HDF5ShardWriter:
#     def __init__(self, output_dir, drive_dir, shard_size=500):
#         self.output_dir = output_dir
#         self.drive_dir = drive_dir
#         self.shard_size = shard_size
#         self.shard_count = 0
#         self.buffer_img = []
#         self.buffer_mask = []

#     def add(self, img, mask):
#         self.buffer_img.append(img)
#         self.buffer_mask.append(mask)

#         # Flush significantly more often to prevent RAM spikes
#         if len(self.buffer_img) >= self.shard_size:
#             self.flush()

#     def flush(self):
#         if not self.buffer_img: return

#         shard_name = f"shard_{self.shard_count:03d}.h5"
#         local_path = os.path.join(self.output_dir, shard_name)
#         drive_path = os.path.join(self.drive_dir, shard_name)

#         print(f"   üíæ Writing {shard_name} ({len(self.buffer_img)} chips)...")

#         try:
#             # Create arrays one by one to manage memory
#             imgs_arr = np.array(self.buffer_img, dtype=np.float16)
#             self.buffer_img = [] # Clear list immediately

#             masks_arr = np.array(self.buffer_mask, dtype=np.uint8)
#             self.buffer_mask = [] # Clear list immediately

#             # Write to HDF5
#             with h5py.File(local_path, 'w') as f:
#                 f.create_dataset("images", data=imgs_arr, compression="lzf")
#                 f.create_dataset("masks", data=masks_arr, compression="lzf")

#             # Clear Numpy Arrays immediately
#             del imgs_arr, masks_arr
#             gc.collect()

#             print(f"   ‚òÅÔ∏è Moving to Drive: {drive_path}")
#             shutil.move(local_path, drive_path)

#         except Exception as e:
#             print(f"‚ùå Error writing shard: {e}")
#             self.buffer_img = []
#             self.buffer_mask = []
#             gc.collect()

#         self.shard_count += 1

# def create_training_data():
#     if len(glob.glob(f"{DRIVE_BACKUP_DIR}/*.h5")) > 0:
#         print("‚úÖ Backup found on Drive. Skipping generation.")
#         return

#     all_files = sorted(glob.glob(f"{INPUT_FOLDER}/*.tif"))
#     district_files = [f for f in all_files if DISTRICT_TARGET.lower() in f.lower()]

#     if len(district_files) < 2: return

#     print(f"üìç Found {len(district_files)} images. Generating HDF5 Shards (ALL PIXELS)...")

#     # Initialize Writer
#     writer = HDF5ShardWriter(TEMP_OUTPUT_DIR, DRIVE_BACKUP_DIR, shard_size=500)

#     total_chips = 0

#     for i in range(len(district_files) - 1):
#         file1 = district_files[i]
#         file2 = district_files[i+1]
#         print(f"üî® Pair: {os.path.basename(file1)} -> {os.path.basename(file2)}")

#         with rasterio.open(file1) as src1, rasterio.open(file2) as src2:
#             height, width = src1.height, src1.width
#             has_swir = src1.count >= 6

#             # Iterate strips
#             for y_offset in tqdm(range(0, height, PROCESS_CHUNK_ROW_SIZE)):
#                 window = Window(0, y_offset, width, min(PROCESS_CHUNK_ROW_SIZE, height - y_offset))

#                 # Read strip
#                 p1_strip = src1.read(window=window)
#                 p2_strip = src2.read(window=window)

#                 if np.max(p1_strip) == 0:
#                     del p1_strip, p2_strip
#                     continue

#                 # Calculations
#                 ndvi1 = calculate_index(p1_strip[3], p1_strip[2])
#                 ndvi2 = calculate_index(p2_strip[3], p2_strip[2])
#                 diff_ndvi = ndvi2 - ndvi1

#                 label_mask = np.zeros(diff_ndvi.shape, dtype=np.uint8)
#                 label_mask[diff_ndvi < -0.15] = 1
#                 label_mask[diff_ndvi > 0.15] = 2

#                 if has_swir:
#                     ndbi1 = calculate_index(p1_strip[4], p1_strip[3])
#                     ndbi2 = calculate_index(p2_strip[4], p2_strip[3])
#                     mndwi1 = calculate_index(p1_strip[1], p1_strip[4])
#                     mndwi2 = calculate_index(p2_strip[1], p2_strip[4])

#                     diff_ndbi = ndbi2 - ndbi1
#                     diff_mndwi = mndwi2 - mndwi1

#                     label_mask[(diff_ndbi > 0.15) & (ndbi1 < 0.1)] = 3
#                     label_mask[(diff_mndwi < -0.20) & (mndwi1 > 0)] = 4
#                     label_mask[(diff_mndwi > 0.20) & (mndwi2 > 0)] = 5

#                 # Slicing Loop
#                 h_s, w_s = p1_strip.shape[1], p1_strip.shape[2]
#                 for y in range(0, h_s - CHIP_SIZE, STRIDE):
#                     for x in range(0, w_s - CHIP_SIZE, STRIDE):
#                         mask_patch = label_mask[y:y+CHIP_SIZE, x:x+CHIP_SIZE]

#                         # --- MODIFIED LOGIC: Save Everything ---
#                         # We only check if the chip is not empty black padding.
#                         # If the max value in the chip is > 0, it contains valid data.

#                         chip_data = p1_strip[:, y:y+CHIP_SIZE, x:x+CHIP_SIZE]

#                         if np.max(chip_data) > 0:
#                             chip1 = chip_data
#                             chip2 = p2_strip[:, y:y+CHIP_SIZE, x:x+CHIP_SIZE]

#                             if not has_swir:
#                                 pad = np.zeros((2, CHIP_SIZE, CHIP_SIZE), dtype=chip1.dtype)
#                                 chip1 = np.concatenate((chip1, pad))
#                                 chip2 = np.concatenate((chip2, pad))

#                             combined = np.concatenate((chip1, chip2)).astype(np.float16)
#                             writer.add(combined, mask_patch)
#                             total_chips += 1

#                 # Cleanup
#                 del p1_strip, p2_strip, label_mask, ndvi1, ndvi2
#                 if has_swir: del ndbi1, ndbi2, mndwi1, mndwi2
#                 gc.collect()

#     writer.flush()
#     print(f"‚úÖ Done! {total_chips} chips saved to HDF5 shards on Drive.")

# create_training_data()


# import rasterio
# from rasterio.windows import Window
# import numpy as np
# import glob
# import os
# import random
# from tqdm import tqdm
# import h5py
# import shutil
# import gc

# # --- CONFIGURATION ---
# CHIP_SIZE = 256
# STRIDE = 128
# DISTRICT_TARGET = "Ahmadabad"
# PROCESS_CHUNK_ROW_SIZE = 1024
# INPUT_FOLDER = "/content/drive/MyDrive/Gujarat_Satellite_Data"
# TEMP_OUTPUT_DIR = "/content/temp_shards"
# DRIVE_BACKUP_DIR = "/content/drive/MyDrive/ahmedabad_hdf5_shards"

# os.makedirs(TEMP_OUTPUT_DIR, exist_ok=True)
# os.makedirs(DRIVE_BACKUP_DIR, exist_ok=True)

# def calculate_index(b1, b2):
#     with np.errstate(divide='ignore', invalid='ignore'):
#         idx = (b1 - b2) / (b1 + b2 + 1e-8)
#     return np.nan_to_num(idx, nan=-1.0)

# class HDF5ShardWriter:
#     def __init__(self, output_dir, drive_dir, shard_size=500):
#         self.output_dir = output_dir
#         self.drive_dir = drive_dir
#         self.shard_size = shard_size
#         self.shard_count = 0
#         self.buffer_img = []
#         self.buffer_mask = []

#     def add(self, img, mask):
#         self.buffer_img.append(img)
#         self.buffer_mask.append(mask)

#         # Flush significantly more often to prevent RAM spikes
#         if len(self.buffer_img) >= self.shard_size:
#             self.flush()

#     def flush(self):
#         if not self.buffer_img: return

#         shard_name = f"shard_{self.shard_count:03d}.h5"
#         local_path = os.path.join(self.output_dir, shard_name)
#         drive_path = os.path.join(self.drive_dir, shard_name)

#         print(f"   üíæ Writing {shard_name} ({len(self.buffer_img)} chips)...")

#         try:
#             # Create arrays one by one to manage memory
#             imgs_arr = np.array(self.buffer_img, dtype=np.float16)
#             self.buffer_img = [] # Clear list immediately

#             masks_arr = np.array(self.buffer_mask, dtype=np.uint8)
#             self.buffer_mask = [] # Clear list immediately

#             # Write to HDF5 ('w' mode overwrites existing files)
#             with h5py.File(local_path, 'w') as f:
#                 f.create_dataset("images", data=imgs_arr, compression="lzf")
#                 f.create_dataset("masks", data=masks_arr, compression="lzf")

#             # Clear Numpy Arrays immediately
#             del imgs_arr, masks_arr
#             gc.collect()

#             print(f"   ‚òÅÔ∏è Overwriting to Drive: {drive_path}")
#             shutil.move(local_path, drive_path)

#         except Exception as e:
#             print(f"‚ùå Error writing shard: {e}")
#             self.buffer_img = []
#             self.buffer_mask = []
#             gc.collect()

#         self.shard_count += 1

# def create_training_data():
#     # --- CHANGE: Removed the "Backup Found" check block here ---
#     # The code will now proceed directly to generation.

#     all_files = sorted(glob.glob(f"{INPUT_FOLDER}/*.tif"))
#     district_files = [f for f in all_files if DISTRICT_TARGET.lower() in f.lower()]

#     if len(district_files) < 2: return

#     print(f"üìç Found {len(district_files)} images. RE-GENERATING HDF5 Shards (ALL PIXELS)...")
#     print(f"‚ö†Ô∏è Warning: This will OVERWRITE existing shards in {DRIVE_BACKUP_DIR}")

#     # Initialize Writer
#     writer = HDF5ShardWriter(TEMP_OUTPUT_DIR, DRIVE_BACKUP_DIR, shard_size=500)

#     total_chips = 0

#     for i in range(len(district_files) - 1):
#         file1 = district_files[i]
#         file2 = district_files[i+1]
#         print(f"üî® Pair: {os.path.basename(file1)} -> {os.path.basename(file2)}")

#         with rasterio.open(file1) as src1, rasterio.open(file2) as src2:
#             height, width = src1.height, src1.width
#             has_swir = src1.count >= 6

#             # Iterate strips
#             for y_offset in tqdm(range(0, height, PROCESS_CHUNK_ROW_SIZE)):
#                 window = Window(0, y_offset, width, min(PROCESS_CHUNK_ROW_SIZE, height - y_offset))

#                 # Read strip
#                 p1_strip = src1.read(window=window)
#                 p2_strip = src2.read(window=window)

#                 if np.max(p1_strip) == 0:
#                     del p1_strip, p2_strip
#                     continue

#                 # Calculations
#                 ndvi1 = calculate_index(p1_strip[3], p1_strip[2])
#                 ndvi2 = calculate_index(p2_strip[3], p2_strip[2])
#                 diff_ndvi = ndvi2 - ndvi1

#                 label_mask = np.zeros(diff_ndvi.shape, dtype=np.uint8)
#                 label_mask[diff_ndvi < -0.15] = 1
#                 label_mask[diff_ndvi > 0.15] = 2

#                 if has_swir:
#                     ndbi1 = calculate_index(p1_strip[4], p1_strip[3])
#                     ndbi2 = calculate_index(p2_strip[4], p2_strip[3])
#                     mndwi1 = calculate_index(p1_strip[1], p1_strip[4])
#                     mndwi2 = calculate_index(p2_strip[1], p2_strip[4])

#                     diff_ndbi = ndbi2 - ndbi1
#                     diff_mndwi = mndwi2 - mndwi1

#                     label_mask[(diff_ndbi > 0.15) & (ndbi1 < 0.1)] = 3
#                     label_mask[(diff_mndwi < -0.20) & (mndwi1 > 0)] = 4
#                     label_mask[(diff_mndwi > 0.20) & (mndwi2 > 0)] = 5

#                 # Slicing Loop
#                 h_s, w_s = p1_strip.shape[1], p1_strip.shape[2]
#                 for y in range(0, h_s - CHIP_SIZE, STRIDE):
#                     for x in range(0, w_s - CHIP_SIZE, STRIDE):
#                         mask_patch = label_mask[y:y+CHIP_SIZE, x:x+CHIP_SIZE]

#                         # --- LOGIC: Save Everything Valid ---
#                         chip_data = p1_strip[:, y:y+CHIP_SIZE, x:x+CHIP_SIZE]

#                         if np.max(chip_data) > 0:
#                             chip1 = chip_data
#                             chip2 = p2_strip[:, y:y+CHIP_SIZE, x:x+CHIP_SIZE]

#                             if not has_swir:
#                                 pad = np.zeros((2, CHIP_SIZE, CHIP_SIZE), dtype=chip1.dtype)
#                                 chip1 = np.concatenate((chip1, pad))
#                                 chip2 = np.concatenate((chip2, pad))

#                             combined = np.concatenate((chip1, chip2)).astype(np.float16)
#                             writer.add(combined, mask_patch)
#                             total_chips += 1

#                 # Cleanup
#                 del p1_strip, p2_strip, label_mask, ndvi1, ndvi2
#                 if has_swir: del ndbi1, ndbi2, mndwi1, mndwi2
#                 gc.collect()

#     writer.flush()
#     print(f"‚úÖ Done! {total_chips} chips saved to HDF5 shards on Drive.")

# create_training_data()


import rasterio
from rasterio.windows import Window
import numpy as np
import glob
import os
import random
from tqdm import tqdm
import h5py
import shutil
import gc

# --- CONFIGURATION ---
CHIP_SIZE = 256
STRIDE = 128
DISTRICT_TARGET = "Ahmadabad"
PROCESS_CHUNK_ROW_SIZE = 1024
INPUT_FOLDER = "/content/drive/MyDrive/Gujarat_Satellite_Data"
TEMP_OUTPUT_DIR = "/content/temp_shards"
DRIVE_BACKUP_DIR = "/content/drive/MyDrive/ahmedabad_hdf5_shards"

os.makedirs(TEMP_OUTPUT_DIR, exist_ok=True)
os.makedirs(DRIVE_BACKUP_DIR, exist_ok=True)

def calculate_index(b1, b2):
    with np.errstate(divide='ignore', invalid='ignore'):
        idx = (b1 - b2) / (b1 + b2 + 1e-8)
    return np.nan_to_num(idx, nan=-1.0)

class HDF5ShardWriter:
    def __init__(self, output_dir, drive_dir, shard_size=500):
        self.output_dir = output_dir
        self.drive_dir = drive_dir
        self.shard_size = shard_size
        self.shard_count = 0
        self.buffer_img = []
        self.buffer_mask = []

    def add(self, img, mask):
        self.buffer_img.append(img)
        self.buffer_mask.append(mask)

        # Flush significantly more often to prevent RAM spikes
        if len(self.buffer_img) >= self.shard_size:
            self.flush()

    def flush(self):
        if not self.buffer_img: return

        shard_name = f"shard_{self.shard_count:03d}.h5"
        local_path = os.path.join(self.output_dir, shard_name)
        drive_path = os.path.join(self.drive_dir, shard_name)

        print(f"   üíæ Writing {shard_name} ({len(self.buffer_img)} chips)...")

        try:
            # Create arrays one by one to manage memory
            imgs_arr = np.array(self.buffer_img, dtype=np.float16)
            self.buffer_img = [] # Clear list immediately

            masks_arr = np.array(self.buffer_mask, dtype=np.uint8)
            self.buffer_mask = [] # Clear list immediately

            # Write to HDF5 ('w' mode overwrites existing files)
            with h5py.File(local_path, 'w') as f:
                f.create_dataset("images", data=imgs_arr, compression="lzf")
                f.create_dataset("masks", data=masks_arr, compression="lzf")

            # Clear Numpy Arrays immediately
            del imgs_arr, masks_arr
            gc.collect()

            print(f"   ‚òÅÔ∏è Moving to Drive: {drive_path}")
            shutil.move(local_path, drive_path)

        except Exception as e:
            print(f"‚ùå Error writing shard: {e}")
            self.buffer_img = []
            self.buffer_mask = []
            gc.collect()

        self.shard_count += 1

def create_training_data():
    # --- 1. Check for Existing Backup ---
    existing_shards = glob.glob(f"{DRIVE_BACKUP_DIR}/*.h5")

    if len(existing_shards) > 0:
        print(f"‚ö†Ô∏è Found {len(existing_shards)} existing HDF5 shards in backup folder.")
        # --- 2. Prompt User ---
        print("üëâ Enter 'r' to RECREATE chips (Overwrites backup).")
        print("üëâ Enter 'u' to USE existing chips (Skips generation).")
        user_choice = input("Your choice [r/u]: ").strip().lower()

        # --- 3. Handle Choice ---
        if user_choice == 'u':
            print("‚úÖ Selected: Use existing chips. Skipping generation.")
            return
        elif user_choice == 'r':
            print("üîÑ Selected: Recreate chips. Cleaning old backup...")
            # Optional: Delete old files to ensure clean write
            for f in existing_shards:
                try: os.remove(f)
                except: pass
        else:
            print("‚ùå Invalid input. Aborting to be safe.")
            return
    else:
        print("fw No backup found. Starting fresh generation...")

    # --- 4. Generation Logic ---
    all_files = sorted(glob.glob(f"{INPUT_FOLDER}/*.tif"))
    district_files = [f for f in all_files if DISTRICT_TARGET.lower() in f.lower()]

    if len(district_files) < 2:
        print("‚ùå Not enough images found.")
        return

    print(f"üìç Found {len(district_files)} images. Generating HDF5 Shards (ALL PIXELS)...")

    # Initialize Writer
    writer = HDF5ShardWriter(TEMP_OUTPUT_DIR, DRIVE_BACKUP_DIR, shard_size=500)

    total_chips = 0

    for i in range(len(district_files) - 1):
        file1 = district_files[i]
        file2 = district_files[i+1]
        print(f"üî® Pair: {os.path.basename(file1)} -> {os.path.basename(file2)}")

        with rasterio.open(file1) as src1, rasterio.open(file2) as src2:
            height, width = src1.height, src1.width
            has_swir = src1.count >= 6

            # Iterate strips
            for y_offset in tqdm(range(0, height, PROCESS_CHUNK_ROW_SIZE)):
                window = Window(0, y_offset, width, min(PROCESS_CHUNK_ROW_SIZE, height - y_offset))

                # Read strip
                p1_strip = src1.read(window=window)
                p2_strip = src2.read(window=window)

                if np.max(p1_strip) == 0:
                    del p1_strip, p2_strip
                    continue

                # Calculations
                ndvi1 = calculate_index(p1_strip[3], p1_strip[2])
                ndvi2 = calculate_index(p2_strip[3], p2_strip[2])
                diff_ndvi = ndvi2 - ndvi1

                label_mask = np.zeros(diff_ndvi.shape, dtype=np.uint8)
                label_mask[diff_ndvi < -0.15] = 1
                label_mask[diff_ndvi > 0.15] = 2

                if has_swir:
                    ndbi1 = calculate_index(p1_strip[4], p1_strip[3])
                    ndbi2 = calculate_index(p2_strip[4], p2_strip[3])
                    mndwi1 = calculate_index(p1_strip[1], p1_strip[4])
                    mndwi2 = calculate_index(p2_strip[1], p2_strip[4])

                    diff_ndbi = ndbi2 - ndbi1
                    diff_mndwi = mndwi2 - mndwi1

                    label_mask[(diff_ndbi > 0.15) & (ndbi1 < 0.1)] = 3
                    label_mask[(diff_mndwi < -0.20) & (mndwi1 > 0)] = 4
                    label_mask[(diff_mndwi > 0.20) & (mndwi2 > 0)] = 5

                # Slicing Loop
                h_s, w_s = p1_strip.shape[1], p1_strip.shape[2]
                for y in range(0, h_s - CHIP_SIZE, STRIDE):
                    for x in range(0, w_s - CHIP_SIZE, STRIDE):
                        mask_patch = label_mask[y:y+CHIP_SIZE, x:x+CHIP_SIZE]

                        # --- LOGIC: Save Everything Valid ---
                        chip_data = p1_strip[:, y:y+CHIP_SIZE, x:x+CHIP_SIZE]

                        if np.max(chip_data) > 0:
                            chip1 = chip_data
                            chip2 = p2_strip[:, y:y+CHIP_SIZE, x:x+CHIP_SIZE]

                            if not has_swir:
                                pad = np.zeros((2, CHIP_SIZE, CHIP_SIZE), dtype=chip1.dtype)
                                chip1 = np.concatenate((chip1, pad))
                                chip2 = np.concatenate((chip2, pad))

                            combined = np.concatenate((chip1, chip2)).astype(np.float16)
                            writer.add(combined, mask_patch)
                            total_chips += 1

                # Cleanup
                del p1_strip, p2_strip, label_mask, ndvi1, ndvi2
                if has_swir: del ndbi1, ndbi2, mndwi1, mndwi2
                gc.collect()

    writer.flush()
    print(f"‚úÖ Done! {total_chips} chips saved to HDF5 shards on Drive.")

create_training_data()
```

```python
import torch
import numpy as np
import glob
import os
import shutil
import h5py
import json
import random
from torch.utils.data import Dataset, DataLoader, Sampler

# --- 1. SETUP DATA ---
DRIVE_BACKUP_DIR = "/content/drive/MyDrive/ahmedabad_hdf5_shards"
LOCAL_DATA_DIR = "/content/train_data_shards"

def restore_shards():
    if os.path.exists(LOCAL_DATA_DIR) and len(os.listdir(LOCAL_DATA_DIR)) > 0:
        print("‚úÖ Data already present locally.")
        return
    os.makedirs(LOCAL_DATA_DIR, exist_ok=True)
    shards = glob.glob(f"{DRIVE_BACKUP_DIR}/*.h5")
    if len(shards) == 0:
        print("‚ùå No shards found. Run Section 2 first.")
        return
    print(f"üì¶ Copying {len(shards)} shards to local disk...")
    for shard in shards:
        shutil.copy(shard, LOCAL_DATA_DIR)
    print("‚úÖ Restore Complete.")

restore_shards()

# --- 2. DATASET CLASS ---
class HDF5Dataset(Dataset):
    def __init__(self, file_list):
        self.shard_paths = sorted(file_list)
        self.index_map = []
        self.files = {}
        self.shard_sizes = []

        print(f"   ‚öôÔ∏è Indexing {len(self.shard_paths)} shards...")

        # Build Index
        for shard_idx, p in enumerate(self.shard_paths):
            try:
                with h5py.File(p, 'r') as f:
                    num = f['images'].shape[0]
                    self.shard_sizes.append(num)
                    self.index_map.extend([(shard_idx, i) for i in range(num)])
            except:
                print(f"   ‚ö†Ô∏è Warning: Could not read {p}")
                self.shard_sizes.append(0)

    def __len__(self):
        return len(self.index_map)

    def __getitem__(self, idx):
        shard_idx, local_idx = self.index_map[idx]

        # Open file if not open (Keep only 1 file open to save RAM/Time)
        if shard_idx not in self.files:
            if len(self.files) > 0:
                key_to_remove = next(iter(self.files))
                self.files[key_to_remove].close()
                del self.files[key_to_remove]
            self.files[shard_idx] = h5py.File(self.shard_paths[shard_idx], 'r')

        f = self.files[shard_idx]

        # Read Data
        image = f['images'][local_idx]
        mask = f['masks'][local_idx]

        # Normalize
        return (torch.from_numpy(image.astype(np.float32) / 3000.0),
                torch.from_numpy(mask.astype(np.longlong)))

# --- 3. SMART SAMPLER (The Speed Fix) ---
class BlockShuffleSampler(Sampler):
    def __init__(self, dataset):
        self.dataset = dataset

    def __iter__(self):
        # 1. Create chunks of indices corresponding to each shard
        shard_indices = {}
        global_idx = 0
        for shard_id, count in enumerate(self.dataset.shard_sizes):
            # Indices for this specific file
            indices = list(range(global_idx, global_idx + count))
            # Shuffle indices WITHIN the file (Randomness)
            np.random.shuffle(indices)
            shard_indices[shard_id] = indices
            global_idx += count

        # 2. Shuffle the ORDER of files (Data Mixing)
        shard_order = list(shard_indices.keys())
        np.random.shuffle(shard_order)

        # 3. Flatten into one stream
        final_order = []
        for shard_id in shard_order:
            final_order.extend(shard_indices[shard_id])

        return iter(final_order)

    def __len__(self):
        return len(self.dataset)

# --- 4. INITIALIZE LOADERS ---
all_shards = sorted(glob.glob(f"{LOCAL_DATA_DIR}/*.h5"))
random.shuffle(all_shards) # Shuffle once so split is random but file-based

# 80/20 Split by FILE (Not by image)
split_idx = int(0.8 * len(all_shards))
train_files = all_shards[:split_idx]
val_files = all_shards[split_idx:]

print(f"üìÇ Training on {len(train_files)} files | Validation on {len(val_files)} files")

# Create independent datasets
train_dataset = HDF5Dataset(train_files)
val_dataset = HDF5Dataset(val_files)

# Loaders
train_loader = DataLoader(
    train_dataset,
    batch_size=32,          # Reduced from 32 to save RAM
    sampler=BlockShuffleSampler(train_dataset),
    num_workers=0,          # <--- CRITICAL FIX: Set to 0
    pin_memory=True
)

val_loader = DataLoader(
    val_dataset,
    batch_size=32,          # Reduced from 32
    shuffle=False,
    num_workers=0,          # <--- CRITICAL FIX: Set to 0
    pin_memory=True
)

print("‚úÖ Loaders Ready (Single-Process Mode for Stability).")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import segmentation_models_pytorch as smp
import numpy as np
import h5py
import glob
import os
import random
from tqdm import tqdm
import torch.nn.functional as F
import gc

# ==========================================
# 1. CONFIGURATION
# ==========================================
CONFIG = {
    "DEVICE": "cuda" if torch.cuda.is_available() else "cpu",
    "NUM_CLASSES": 6,
    "IN_CHANNELS": 12,
    "IMG_SIZE": 128,
    "BATCH_SIZE": 32,
    "MAX_SAMPLES": 8000,  # Limit dataset to 8k samples
    "LR": 5e-4,
    "EPOCHS": 30,
    "SHARDS_DIR": "/content/train_data_shards",
    "MEMMAP_IMG": "/content/train_images.dat",
    "MEMMAP_MSK": "/content/train_masks.dat",
    "MODEL_SAVE_PATH": "/content/drive/MyDrive/Gujarat_Change_Detection_Memmap.pth"
}

# ==========================================
# 2. DISK-BACKED PRE-PROCESSING (Low RAM)
# ==========================================
def prepare_memmap_dataset():
    if os.path.exists(CONFIG["MEMMAP_IMG"]):
        print("‚úÖ Memmap files already exist. Skipping generation.")
        return

    print(f"üíø Creating Disk-Backed Dataset ({CONFIG['MAX_SAMPLES']} samples)...")

    # 1. Initialize empty files on disk (Virtual Memory)
    # Shape: (N, C, H, W) for images, (N, H, W) for masks
    fp_img = np.memmap(CONFIG["MEMMAP_IMG"], dtype='float32', mode='w+',
                       shape=(CONFIG["MAX_SAMPLES"], 12, CONFIG["IMG_SIZE"], CONFIG["IMG_SIZE"]))
    fp_msk = np.memmap(CONFIG["MEMMAP_MSK"], dtype='int64', mode='w+',
                       shape=(CONFIG["MAX_SAMPLES"], CONFIG["IMG_SIZE"], CONFIG["IMG_SIZE"]))

    all_files = sorted(glob.glob(f"{CONFIG['SHARDS_DIR']}/*.h5"))
    if not all_files: raise FileNotFoundError("Run 'Restore Data' step first!")
    random.shuffle(all_files)

    idx = 0
    pbar = tqdm(total=CONFIG['MAX_SAMPLES'])

    for path in all_files:
        if idx >= CONFIG["MAX_SAMPLES"]: break

        try:
            with h5py.File(path, 'r') as f:
                # Read specific indices to avoid loading whole file
                total_in_file = f['images'].shape[0]
                # Take up to 100 images per file
                take = min(total_in_file, 100)
                if take == 0: continue

                # Read raw data
                raw_imgs = f['images'][:take] # (N, 256, 256, 12)
                raw_msks = f['masks'][:take]  # (N, 256, 256)

                # Process ONE BY ONE to keep RAM flat
                for i in range(take):
                    if idx >= CONFIG["MAX_SAMPLES"]: break

                    # 1. To Torch & Normalize
                    img = torch.from_numpy(raw_imgs[i].astype(np.float32) / 3000.0) # (256, 256, 12)
                    msk = torch.from_numpy(raw_msks[i].astype(np.longlong))         # (256, 256)

                    # 2. Permute to (C, H, W)
                    img = img.permute(2, 0, 1)

                    # 3. Resize (Downscale to 128x128)
                    img = F.interpolate(img.unsqueeze(0), size=(CONFIG['IMG_SIZE'], CONFIG['IMG_SIZE']),
                                      mode='bilinear', align_corners=False).squeeze(0)

                    msk = F.interpolate(msk.float().unsqueeze(0).unsqueeze(0), size=(CONFIG['IMG_SIZE'], CONFIG['IMG_SIZE']),
                                      mode='nearest').long().squeeze(0).squeeze(0)

                    # 4. Write to Disk
                    fp_img[idx] = img.numpy()
                    fp_msk[idx] = msk.numpy()

                    idx += 1
                    pbar.update(1)
        except Exception as e:
            # print(f"Error reading {path}: {e}")
            pass

    # Flush changes to disk
    fp_img.flush()
    fp_msk.flush()
    pbar.close()
    print(f"‚úÖ Dataset Created on Disk. Size: {idx} samples.")

# Run the prep
prepare_memmap_dataset()

# ==========================================
# 3. MEMMAP DATASET CLASS
# ==========================================
class MemmapDataset(Dataset):
    def __init__(self):
        # Open files in read-only mode
        self.images = np.memmap(CONFIG["MEMMAP_IMG"], dtype='float32', mode='r',
                                shape=(CONFIG["MAX_SAMPLES"], 12, CONFIG["IMG_SIZE"], CONFIG["IMG_SIZE"]))
        self.masks = np.memmap(CONFIG["MEMMAP_MSK"], dtype='int64', mode='r',
                               shape=(CONFIG["MAX_SAMPLES"], CONFIG["IMG_SIZE"], CONFIG["IMG_SIZE"]))

    def __len__(self):
        return CONFIG["MAX_SAMPLES"]

    def __getitem__(self, idx):
        # Read from disk (OS caches this in RAM automatically)
        img = torch.from_numpy(self.images[idx]) # Already (12, 128, 128)
        mask = torch.from_numpy(self.masks[idx])
        return img, mask

# Create Loader
train_ds = MemmapDataset()
train_loader = DataLoader(train_ds, batch_size=CONFIG["BATCH_SIZE"], shuffle=True, num_workers=2)

# ==========================================
# 4. TRAINING LOOP
# ==========================================
print("üèóÔ∏è Building Model (MiT-B0)...")
model = smp.Segformer(
    encoder_name="mit_b0",
    encoder_weights="imagenet",
    in_channels=CONFIG["IN_CHANNELS"],
    classes=CONFIG["NUM_CLASSES"]
).to(CONFIG["DEVICE"])

optimizer = optim.AdamW(model.parameters(), lr=CONFIG["LR"])
scaler = torch.amp.GradScaler('cuda')
criterion = nn.CrossEntropyLoss()

print(f"üöÄ Training Started...")

for epoch in range(CONFIG["EPOCHS"]):
    model.train()
    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG['EPOCHS']}")

    for img, mask in loop:
        img = img.to(CONFIG["DEVICE"], non_blocking=True)
        mask = mask.to(CONFIG["DEVICE"], non_blocking=True)

        optimizer.zero_grad(set_to_none=True)

        with torch.amp.autocast('cuda'):
            pred = model(img)
            loss = criterion(pred, mask)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        loop.set_postfix(loss=loss.item())

    if (epoch+1) % 5 == 0:
        torch.save(model.state_dict(), CONFIG["MODEL_SAVE_PATH"])
        print(f"   üíæ Checkpoint Saved")

print("üèÅ DONE!")
```

```python
import torch
import numpy as np
import rasterio
from rasterio.windows import Window
from rasterio.enums import Resampling
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import os
import segmentation_models_pytorch as smp
from tqdm import tqdm
import cv2

MODEL_PATH = "/content/drive/MyDrive/ahmedabad_segformer_6class.pth"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# 0=White, 1=Red(Defor), 2=Green(Affor), 3=Yellow(Urb), 4=Cyan(WtrLoss), 5=Blue(WtrGain)
CLASSES = ['Stable', 'Veg Loss', 'Veg Gain', 'Urban Gain', 'Water Loss', 'Water Gain']
CMAP = ListedColormap(['white', 'red', 'green', 'yellow', 'cyan', 'blue'])

inference_model = smp.Segformer(encoder_name="mit_b3", in_channels=12, classes=6)

if os.path.exists(MODEL_PATH):
    inference_model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
    inference_model.to(DEVICE)
    inference_model.eval()
    print(f"‚úÖ Model Loaded from {MODEL_PATH}")
else:
    raise FileNotFoundError(f"‚ùå Model not found at {MODEL_PATH}")

def predict_change_safe(path_before, path_after, model=inference_model, batch_size=32):

    with rasterio.open(path_before) as src1, rasterio.open(path_after) as src2:
        h, w = src1.height, src1.width
        num_bands = src1.count
        print(f"üîç Processing {w}x{h} pixels | Input Bands: {num_bands}...")

        full_prediction = np.zeros((h, w), dtype=np.uint8)

        chip_size = 256
        stride = 256

        batch_tensors = []
        batch_locations = []

        windows = []
        for y in range(0, h, stride):
            for x in range(0, w, stride):
                windows.append((y, x))

        for i, (y, x) in tqdm(enumerate(windows), total=len(windows), desc="üöÄ Inferring"):
            window_h = min(chip_size, h - y)
            window_w = min(chip_size, w - x)
            window = Window(x, y, window_w, window_h)

            img1 = src1.read(window=window)
            img2 = src2.read(window=window)

            pad_h = chip_size - window_h
            pad_w = chip_size - window_w
            if pad_h > 0 or pad_w > 0:
                img1 = np.pad(img1, ((0,0), (0, pad_h), (0, pad_w)), mode='constant')
                img2 = np.pad(img2, ((0,0), (0, pad_h), (0, pad_w)), mode='constant')

            if num_bands == 4:
                dummy1 = np.zeros((2, 256, 256), dtype=img1.dtype)
                dummy2 = np.zeros((2, 256, 256), dtype=img2.dtype)
                img1 = np.concatenate((img1, dummy1), axis=0)
                img2 = np.concatenate((img2, dummy2), axis=0)

            img1 = np.clip(img1, 0, 3000) / 3000.0
            img2 = np.clip(img2, 0, 3000) / 3000.0
            combined = np.concatenate((img1, img2), axis=0).astype(np.float32)

            batch_tensors.append(combined)
            batch_locations.append((y, x, window_h, window_w))

            if len(batch_tensors) == batch_size or i == len(windows) - 1:
                tensor_batch = torch.from_numpy(np.stack(batch_tensors)).to(DEVICE)
                with torch.no_grad():
                    from torch.cuda.amp import autocast
                    with autocast():
                        outputs = model(tensor_batch)
                        preds = torch.argmax(outputs, dim=1).cpu().numpy()

                for j, pred_mask in enumerate(preds):
                    y_loc, x_loc, h_loc, w_loc = batch_locations[j]
                    full_prediction[y_loc:y_loc+h_loc, x_loc:x_loc+w_loc] = pred_mask[:h_loc, :w_loc]

                batch_tensors = []
                batch_locations = []

    return full_prediction

TEST_IMG_BEFORE = "/content/drive/MyDrive/Gujarat_Satellite_Data/Mahesana_2018_10m.tif"
TEST_IMG_AFTER  = "/content/drive/MyDrive/Gujarat_Satellite_Data/Mahesana_2025_10m.tif"

if os.path.exists(TEST_IMG_BEFORE):
    result_map = predict_change_safe(TEST_IMG_BEFORE, TEST_IMG_AFTER)

    print("üé® Visualizing...")
    def get_thumbnail(path, scale_factor=0.05):
        with rasterio.open(path) as src:
            new_h = int(src.height * scale_factor)
            new_w = int(src.width * scale_factor)
            # Try Reading RGB (B4, B3, B2 are indices 2, 1, 0 usually if B2 is 0)
            # Assuming 6-band export: [B2, B3, B4, B8, B11, B12] -> RGB is [2, 1, 0]
            try:
                img = src.read([3, 2, 1], out_shape=(3, new_h, new_w), resampling=Resampling.bilinear)
            except:
                img = src.read([1, 2, 3], out_shape=(3, new_h, new_w), resampling=Resampling.bilinear)
            img = np.clip(img / 3000.0 * 2.5, 0, 1)
            return np.transpose(img, (1, 2, 0))

    h, w = result_map.shape
    small_h, small_w = int(h * 0.05), int(w * 0.05)
    result_small = cv2.resize(result_map, (small_w, small_h), interpolation=cv2.INTER_NEAREST)

    plt.figure(figsize=(15, 6))
    plt.subplot(1, 3, 1); plt.title("Before"); plt.imshow(get_thumbnail(TEST_IMG_BEFORE)); plt.axis('off')
    plt.subplot(1, 3, 2); plt.title("After"); plt.imshow(get_thumbnail(TEST_IMG_AFTER)); plt.axis('off')
    plt.subplot(1, 3, 3); plt.title("6-Class Change Map"); plt.imshow(result_small, cmap=CMAP, interpolation='nearest'); plt.axis('off')
    plt.show()

    output_tif = "/content/drive/MyDrive/Mahesana_Change_Map_QGIS_6Class.tif"
    print(f"üíæ Saving to: {output_tif}")
    with rasterio.open(TEST_IMG_BEFORE) as src:
        profile = src.profile
        profile.update(count=1, dtype=rasterio.uint8, nodata=0, compress='lzw')
        with rasterio.open(output_tif, 'w', **profile) as dst:
            dst.write(result_map, 1)
            dst.write_colormap(1, {
                0: (0,0,0,0), 1: (255,0,0,255), 2: (0,255,0,255),
                3: (255,255,0,255), 4: (0,255,255,255), 5: (0,0,255,255)
            })
    print("‚úÖ Success!")
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns

def calculate_metrics(loader, model, device):
    print(f"üìä Evaluating on {len(loader.dataset)} chips...")
    total_cm = np.zeros((6, 6), dtype=np.int64)

    with torch.no_grad():
        for images, masks in tqdm(loader, desc="Metrics"):
            images, masks = images.to(device), masks.to(device)
            outputs = model(images)
            preds = torch.argmax(outputs, dim=1)

            batch_cm = confusion_matrix(
                masks.cpu().numpy().flatten(),
                preds.cpu().numpy().flatten(),
                labels=[0, 1, 2, 3, 4, 5]
            )
            total_cm += batch_cm
    return total_cm

conf_matrix = calculate_metrics(val_loader, model, DEVICE)

classes = ['Stable', 'Veg Loss', 'Veg Gain', 'Urban', 'Water Loss', 'Water Gain']
metrics = []

for i in range(6):
    tp = conf_matrix[i, i]
    fp = conf_matrix[:, i].sum() - tp
    fn = conf_matrix[i, :].sum() - tp

    precision = tp / (tp + fp + 1e-10)
    recall = tp / (tp + fn + 1e-10)
    iou = tp / (tp + fp + fn + 1e-10)

    metrics.append({
        "Class": classes[i],
        "Precision": round(precision, 4),
        "Recall": round(recall, 4),
        "IoU": round(iou, 4)
    })

metrics_df = pd.DataFrame(metrics)
print("\nüèÜ FINAL 6-CLASS REPORT üèÜ")
print(metrics_df.to_string(index=False))
print(f"‚≠ê Mean IoU: {metrics_df['IoU'].mean():.4f}")

plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.title('Confusion Matrix (6-Class)')
plt.show()
```

```python
import os
import shutil
import time
from google.colab import drive
from tqdm.notebook import tqdm

def split_dataset_safe():
    # --- 1. CONFIGURATION ---
    SOURCE_DIR = "/content/drive/MyDrive/Gujarat_Satellite_Data"

    DEST_DIRS = {
        "Part1": "/content/drive/MyDrive/Gujarat_Data_Part1", # 2017-2019
        "Part2": "/content/drive/MyDrive/Gujarat_Data_Part2", # 2020-2022
        "Part3": "/content/drive/MyDrive/Gujarat_Data_Part3"  # 2023-2025
    }

    YEAR_GROUPS = {
        "Part1": ["2017", "2018", "2019"],
        "Part2": ["2020", "2021", "2022"],
        "Part3": ["2023", "2024", "2025"]
    }

    # --- 2. SETUP ---
    if not os.path.exists('/content/drive'):
        print("üîå Mounting Google Drive...")
        drive.mount('/content/drive')

    for key, path in DEST_DIRS.items():
        os.makedirs(path, exist_ok=True)

    # --- 3. FILTERING ---
    print(f"üîç Scanning source folder...")
    try:
        all_files = [f for f in os.listdir(SOURCE_DIR) if os.path.isfile(os.path.join(SOURCE_DIR, f))]
    except OSError as e:
        print(f"‚ùå Critical Drive Error: {e}")
        print("üëâ Please go to 'Runtime' > 'Disconnect and delete runtime' and try again.")
        return

    print(f"üì¶ Found {len(all_files)} files.")

    # --- 4. SLOW COPY PROCESS ---
    copied_count = 0
    skipped_count = 0

    pbar = tqdm(all_files, desc="Copying (Throttled)", unit="file")

    for i, filename in enumerate(pbar):
        source_file = os.path.join(SOURCE_DIR, filename)
        target_folder = None

        # Identify Year Group
        for part, years in YEAR_GROUPS.items():
            if any(year in filename for year in years):
                target_folder = DEST_DIRS[part]
                break

        if target_folder is None: continue

        target_file = os.path.join(target_folder, filename)

        # Skip if exists (Resume capability)
        if os.path.exists(target_file):
            if os.path.getsize(source_file) == os.path.getsize(target_file):
                skipped_count += 1
                continue

        try:
            # COPY
            shutil.copy2(source_file, target_file)

            # FORCE SYNC (Writes data to disk immediately)
            os.sync()

            # SLEEP (Prevents Drive Disconnect)
            time.sleep(3)

            copied_count += 1
            pbar.set_postfix(status=f"Saved to {os.path.basename(target_folder)}")

            # Extended sleep every 10 files
            if i > 0 and i % 10 == 0:
                time.sleep(5)

        except OSError as e:
            if "Transport endpoint is not connected" in str(e):
                print("\n‚ùå CRITICAL: Google Drive Disconnected again.")
                print("üõë Stopping script to prevent file corruption.")
                print("üëâ You must Restart Runtime to fix this.")
                break
            else:
                print(f"\n‚ö†Ô∏è Error copying {filename}: {e}")

    print(f"\n‚úÖ Session Ended. Copied: {copied_count} | Skipped: {skipped_count}")

if __name__ == "__main__":
    split_dataset_safe()
```
